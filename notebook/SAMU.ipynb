```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os

%matplotlib inline
```

```python
#plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (16, 10)
```

```python
def getDataFrame(csv_list, encode_list, sepa=';'):
    if(len(csv_list) != len(encode_list)):
        raise ValueError('Listas com tamanhos diferentes')
    
    result = {}
    for i in range(len(csv_list)):
        name = str(os.path.basename(csv_list[i]))
        encode = encode_list[i]['encoding']
        file_path = csv_list[i]
        #print(name, encode, file_path)
        csv = pd.read_csv(file_path, encoding=encode, sep=sepa)
        result[name] = csv
    
    return result

def getFiles(myPath, extesion='.csv'):
    return [os.path.join(myPath, f) for f in os.listdir(myPath) if f.endswith(extesion)]
```

```python
myPath = '../database/SAMU/converted/'
csv_files = getFiles(myPath)
raw_csvs = {}
for file in csv_files:
        name = str(os.path.basename(file))
        csv = pd.read_csv(file, sep=';')
        raw_csvs[name] = csv

for x in raw_csvs.keys():
    print(raw_csvs[x].head())
    print()
```
